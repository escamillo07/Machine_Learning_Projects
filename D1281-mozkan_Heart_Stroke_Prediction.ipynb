Data vocabulary:

1. #3 (age): age in years:  yaş arttıkça heartsroke riski artmaktadır.

2. #4 (sex): sex (1 = male; 0 = female) :  erkeklerde daha çok görülmektedir.

3. #9 (cp): cp: chest pain type | Value 0: typical angina | Value 1: atypical angina | Value 2: non-anginal pain | Value 3: asymptomatic: belirti vermeyen ağrılar

4. #10 (trestbps): resting blood pressure (in mm Hg on admission to the hospital) : dinlenme esnasında atımlar arasında kan basıncı düşer. buna resting blood pressure denir.

5. #12 (chol): serum cholestoral in mg/dl   :  Bir kişinin serum kolesterol seviyesi, kanındaki toplam kolesterol miktarını temsil eder. Bir kişinin serum kolesterol düzeyi, kandaki yüksek yoğunluklu lipoprotein (HDL), düşük yoğunluklu lipoprotein (LDL) ve trigliserit miktarını içerir. Trigliseritler, kolesterol ile dolu bir yağ türüdür.

6. #16 (fbs): (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)  gece boyunca aç kaldıktan sonra (yemek yemeden) kan şekerinizi ölçer. Açlık kan şekeri seviyesinin 99 mg/dL veya daha düşük olması normaldir, 100 ila 125 mg/dL prediyabetiniz olduğunu gösterir,

7. #19 (restecg): resting electrocardiographic results | Value 0: normal | Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV) | Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria..   EKG: (elektrokardiyogram), dinlenme halindeyken kalbinizin elektriksel aktivitesini kaydeder.

8. #32 (thalach): maximum heart rate achieved : yaşa göre kalp atım sayısı.

9. #38 (exang): exercise induced angina (1 = yes; 0 = no)   anjinin en yaygın şeklidir. Genellikle aktivite (egzersiz) sırasında olur ve dinlenme veya anjina ilaçları ile geçer. Örneğin yokuş çıkarken veya soğuk havada çıkan ağrı anjina olabilir.

10. #40 (oldpeak): ST depression induced by exercise relative to rest: Egzersize bağlı ST segment depresyonu, obstrüktif koroner damar sertliği tanısı için güvenilir bir EKG bulgusu olarak kabul edilir.

11. #41 (slope): the slope of the peak exercise ST segment | Value 1: upsloping | Value 2: flat | Value 3: downsloping  egzersiz sırasında, en yüksek egzersizde maksimum depresyon ile bastırılır. Egzersiz sırasında normal ST segmenti bu nedenle keskin bir şekilde yukarı doğru eğimlidir.

12. #44 (ca): number of major vessels (0-3) colored by flourosopy büyük damar sayısı

13. #51 (thal): 3 = normal; 6 = fixed defect; 7 = reversable defect  A blood disorder called thalassemia...Talasemi genetik (kalıtım yoluyla) geçiş gösteren ve ağır derecede kansızlık nedeni olan bir hastalıktır.

14. #58 (num) (the predicted attribute): Value 0: < 50% diameter narrowing | Value 1: > 50% diameter narrowing kalp hastalığı teşhisi (anjiyografik hastalık durumu)

* https://archive.ics.uci.edu/ml/datasets/heart+disease

# Data Preprocessing

```python
X = df.drop('target',axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import plot_roc_curve, plot_precision_recall_curve, roc_auc_score, auc, roc_curve, average_precision_score, precision_recall_curve,plot_confusion_matrix
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    print("----------------------------------------------------------")
    print("                  Test_Set")
    print("----------------------------------------------------------")
    print(confusion_matrix(y_test, y_pred))
    
    print(classification_report(y_test, y_pred))
    print("----------------------------------------------------------")
    print("                 Train_Set")
    print("----------------------------------------------------------")
    print(confusion_matrix(y_train, y_train_pred))
    
    print(classification_report(y_train, y_train_pred))
```

```python
eval_metric(log_model, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
plot_confusion_matrix(log_model, X_test_scaled, y_test);
```

```python
plot_confusion_matrix(log_model, X_train_scaled, y_train);
```
